```python
#collect product details 
#import requests

#asin_list

url = "https://real-time-amazon-data.p.rapidapi.com/product-details"

headers = {
    "x-rapidapi-key": "e205fd7dcbmsh9458a15a00fb8ebp123bcbjsn4d4f5f353fc0",
    "x-rapidapi-host": "real-time-amazon-data.p.rapidapi.com"
}

# Dictionary to store responses
responses = {}

# Iterate through each ASIN in the list
for asin in asin_list:
    querystring = {
        "asin": asin,
        'product_title': "Apple iPhone 11, 64GB, (PRODUCT)RED - Fully Unlocked (Renewed)",
        'product_description': "adidas womens VL Court 3.0",
        'climate_pledge_friendly': 'false',
        'product_price': "$74.95"
    }

    proddesc_response = requests.get(url, headers=headers, params=querystring)
    
    # Store the response in the dictionary
    responses[asin] = proddesc_response.json()

# Print the responses
for asin, response in responses.items():
    print(f"Response for ASIN {asin}: {response}")
```

```python
# Accessing data from the dictionary
for asin, response in responses.items():
    # Do something with each response
    print(f"ASIN: {asin}, Response: {response}")
```

```python
# Extracting specific information from the responses stored in the dictionary
for asin, response in responses.items():
    product_title = response.get('product_title', 'N/A')
    product_price = response.get('product_price', 'N/A')
    print(f"ASIN: {asin}, Title: {product_title}, Price: {product_price}")
```
