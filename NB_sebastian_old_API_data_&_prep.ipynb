## Test Data - Merging and Cleaning

```python
# Concatenate the review dataframes
all_reviews = pd.concat([all_reviews_1, all_reviews_2, all_reviews_3])
```

```python
print(all_reviews.shape)

# Remove duplicates from the combined dataframe
all_reviews.drop_duplicates(subset=['reviewId'], inplace=True)

print(all_reviews.shape)
```

```python
# List of null-like values to filter out
null_like_values = [None, np.nan, "NULL", "", '', '0']

all_reviews_with_nulls = all_reviews.copy()

# Filter rows where reviewId is in null_like_values
all_reviews = all_reviews[~all_reviews['reviewId'].isin(null_like_values)]

# Optionally, if you want to modify the original DataFrame inplace:
# all_reviews = all_reviews[~all_reviews['reviewId'].isin(null_like_values)]

# Print number of rows removed
rows_removed = len(all_reviews_with_nulls) - len(all_reviews)
print(f"Removed {rows_removed} rows with null-like values in 'reviewId' column.")

# Now 'filtered_reviews' contains rows where 'reviewId' is not null-like
```

```python
# Display the first few rows of the dataframe to inspect the 'date' column
print(all_reviews['date'].head())

# Remove the unnecessary text and keep only the date part
all_reviews['date_cleaned'] = all_reviews['date'].str.replace(r'Reviewed in .* on ', '', regex=True)

# Convert the cleaned date strings to datetime objects
all_reviews['date_cleaned'] = pd.to_datetime(all_reviews['date_cleaned'])

# Verify the conversion by displaying the first few rows
print(all_reviews[['date', 'date_cleaned']].head())
```

```python
# Display the first few rows of the dataframe to inspect the 'rating' column
print(all_reviews['rating'].head())

# Extract the numeric part of the rating and convert it to a float
all_reviews['rating_cleaned'] = all_reviews['rating'].str.extract(r'(\d+\.\d+)').astype(float)

# Verify the conversion by displaying the first few rows
print(all_reviews[['rating', 'rating_cleaned']].head())
```

```python
# Columns to delete
columns_to_delete = ['Unnamed: 0', 'imageUrlList', 'numberOfHelpful', 'parentReviewId', 'date', 'rating']

# Drop the specified columns from the DataFrame
all_reviews.drop(columns=columns_to_delete, inplace=True)
all_reviews
```

```python
# Define the desired order of columns
desired_order = ['reviewId', 'asin', 'date_cleaned', 'username', 'title', 'keyword', 'verified', 'rating_cleaned', 'text']

# Reorder the columns
all_reviews = all_reviews[desired_order]

# Verify the changes
all_reviews
```

```python
# Column that tags whether subset is train or test
all_reviews['train_test'] = 'test'

# Verify the changes
all_reviews
```
