# Master Data Merging

```python
# Import libraries
import pandas as pd
```

# Read in datasets

## Old API Amazon Reviews

```python
# Read in datasets
reviews_old_API_data = pd.read_csv('processed_reviews_old_API_data.csv')
```

```python
# Preview data
reviews_old_API_data
```

## New API Amazon Reviews

```python
# Read in datasets
reviews_new_API_data = pd.read_csv('processed_reviews_new_API_data.csv')
```

```python
# Preview data
reviews_new_API_data
```

## Merging Old and New API Amazon Reviews

```python
# Concatenate the dataframes
all_reviews_old_and_new = pd.concat([reviews_old_API_data, reviews_new_API_data], ignore_index=True)
```

```python
# Preview data
all_reviews_old_and_new
```

```python
# Save dataset
all_reviews_old_and_new.to_csv('test_set.csv', index=False)
print('Saved as CSV.')
```

## Attribute = 'Sustainability'

```python
# Read in datasets
synth_sustainability_positive = pd.read_csv('synthetic_data_sustainability_positive.csv')
synth_sustainability_negative = pd.read_csv('synthetic_data_sustainability_negative.csv')
```

```python
# Change column names and preview data
synth_sustainability_positive.rename(columns={
    'attribute': 'attr_sustainability',
    'attribute_sentiment': 'attr_sentiment_sustainability'}, inplace=True)

synth_sustainability_positive
```

```python
# Change column names and preview data
synth_sustainability_negative.rename(columns={
    'attribute': 'attr_sustainability',
    'attribute_sentiment': 'attr_sentiment_sustainability'}, inplace=True)

synth_sustainability_negative
```

# Training Data Merging

```python
# Randomly select 1500 rows
sampled_reviews = all_reviews_old_and_new.sample(n=3000, random_state=42)

# Keep only the 'text' column
sampled_reviews = sampled_reviews[['text']]

# Rename columns
sampled_reviews.rename(columns={'text': 'amazon_review'}, inplace=True)

# Preview data
sampled_reviews
```

```python
# Add attr columns with neutral values
sampled_reviews['attr_sustainability'] = 0
sampled_reviews['attr_sentiment_sustainability'] = 0

# Preview data
sampled_reviews
```

```python
# Concatenate the dataframes
train_set_sustainability = pd.concat([synth_sustainability_positive, synth_sustainability_negative, sampled_reviews], ignore_index=True)
```

```python
# Preview data
train_set_sustainability
```

```python
# Save dataset
train_set_sustainability.to_csv('train_set_sustainability.csv', index=False)
print('Saved as CSV.')
```
